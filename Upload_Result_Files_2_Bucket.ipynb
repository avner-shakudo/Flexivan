```python


def upload_to_s3(file_path, bucket_name, object_name=None):
    """
    Uploads a file to an S3 bucket.

    Args:
        file_path (str): Local path to the file to upload.
        bucket_name (str): Name of the S3 bucket.
        object_name (str): S3 object name. If None, file_path name is used.

    Returns:
        bool: True if upload succeeded, False otherwise.
    """
    if object_name is None:
        import os
        object_name = os.path.basename(file_path)

    # Create S3 client
    s3 = boto3.client('s3')

    try:
        s3.upload_file(file_path, bucket_name, object_name)
        print(f"File uploaded successfully to s3://{bucket_name}/{object_name}")
        return True
    except FileNotFoundError:
        print("The file was not found")
        return False
    except NoCredentialsError:
        print("Credentials not available")
        return False
    except ClientError as e:
        print("Unexpected error:", e)
        return False

# Example usage

```

```python
Folder = '/Users/avner/flexivan/Daily prediction/DATA'
bucket_name = 'flexivan-shakudo'
```

```python
# Get a DF with filenames to upload

import os

def get_filtered_csv_files(folder_path, substring):
    """
    Returns a list of CSV filenames in folder_path that contain the given substring.

    Parameters:
        folder_path (str): Path to the folder to search in.
        substring (str): Substring to filter filenames.

    Returns:
        List[str]: List of matching CSV filenames (not full paths).
    """
    filtered_files = []
    
    for filename in os.listdir(folder_path):
        # Ensure it's a file
        full_path = os.path.join(folder_path, filename)
        if os.path.isfile(full_path):
            # Filter CSV files containing the substring
            if filename.lower().endswith('.csv') and substring in filename:
                filtered_files.append(filename)
    
    return filtered_files

FILENAMES = get_filtered_csv_files(Folder, 'DETAILED_PREDICTION_RESULTS')

print(f'{len(FILENAMES)} file found')
```

```python
# Upload files to bucket
Target_Folder = '/322/321'

print('Uploading result files to bucket...')

for filename in tqdm(FILENAMES):
    upload_to_s3(f'{Folder}/{filename}', bucket_name, f'{Target_Folder}/{filename}')

print('DONE.')
```
